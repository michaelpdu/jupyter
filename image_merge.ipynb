```python
import tensorflow as tf

c1 = tf.constant([[1.,2.],[3.,4.]])
c2 = tf.constant(2., shape=[2, 2])
c1_large_c2 = tf.math.less(c1,2)
c1_with_mask = tf.boolean_mask(c1, c1_large_c2)
with tf.Session() as sess:
    value = sess.run(c1_large_c2)
    print(value)
    value = sess.run(c1_with_mask)
    print(value)
```

```python
import tensorflow as tf
import numpy as np

c1 = tf.constant([[1.,2.],[3.,4.]])
c2 = tf.constant([[6.,7.],[8.,9.]])
mask = np.array([[True,False],[True,False]])
c1_with_mask = tf.boolean_mask(c1, mask)
with tf.Session() as sess:
    value = sess.run(c1_with_mask)
    print(value)
```

```python
import tensorflow as tf
from skimage import io, color
import numpy as np

img1 = io.imread('./glasses_600_600.jpg')/255.
img2 = io.imread('./xinchi_aligned.png')/255.
print('img1 shape:', img1.shape)
print('img2 shape:', img2.shape)
print(img1.shape[2])

glasses_image = tf.placeholder(tf.float32,shape=[600,600,3], name='glasses_input')
face_image = tf.placeholder(tf.float32,shape=[600,600,3], name='face_input')

merged_img = tf.clip_by_value(tf.where(tf.less(glasses_image,100/255.), glasses_image, face_image), 0, 1.)
with tf.Session() as sess:
    face_with_glasses = sess.run(merged_img,feed_dict={glasses_image:img1, face_image:img2})
    io.imshow(face_with_glasses)
    plt.show()
```

```python
import tensorflow as tf
import numpy as np

sess=tf.Session()

a=np.array([[1,0,0],[0,1,1]])
a1=np.array([[3,2,3],[4,5,6]])

print(sess.run(tf.less(a,1)))
print(sess.run(tf.where(tf.less(a,1),a1,1-a1)))
print(a*a1)

z = np.zeros((3,3), dtype=np.float32)
zero = tf.constant(z)+1
print(sess.run(zero))

#
c=np.array([[0.1,0.9,0],[0.2,1.0,0.95]], dtype=np.float32)
print(c)
c[c>0.9]=1
print(c)
```
